```python
import pandas as pd
df = pd.read_csv("../data/raw/07_scheme_performance.csv")

returns_cols = [
    "return_1y_pct",
    "return_3y_pct",
    "return_5y_pct"
]

for col in returns_cols:
    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )
```

```python
anomalies = df[
    (df["return_1y"] > 100) |
    (df["return_1y"] < -100)
]
```

```python
df = df[
    (df["expense_ratio"] >= 0.1)
    &
    (df["expense_ratio"] <= 2.5)
]
```

```python
df.to_csv(
    "../data/processed/scheme_performance_clean.csv",
    index=False
)
```
